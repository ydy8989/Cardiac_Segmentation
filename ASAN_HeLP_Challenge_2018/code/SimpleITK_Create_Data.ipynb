```python
image = sitk.ReadImage('./image/sample_dia.mha')
print('Origin',image.GetOrigin())
print('Spacing',image.GetSpacing())
print('Direction',image.GetDirection())
```

```python
image.SetSpacing((1.25,1.25,1.25))
print('Spacing',image.GetSpacing())
```

```python
fig = plt.figure()
ax = fig.add_subplot(2,2,1)
ax.imshow(sitk.GetArrayFromImage(image[:,:,10]),cmap='gray')
ax.set_title('z: 10')
ax = fig.add_subplot(2,2,2)
ax.imshow(sitk.GetArrayFromImage(image[:,:,20]),cmap='gray')
ax.set_title('z: 20')
ax = fig.add_subplot(2,2,3)
ax.imshow(sitk.GetArrayFromImage(image[:,:,30]),cmap='gray')
ax.set_title('z: 30')
ax = fig.add_subplot(2,2,4)
ax.imshow(sitk.GetArrayFromImage(image[:,:,40]),cmap='gray')
ax.set_title('z: 40')
plt.subplots_adjust(hspace = 0.5, wspace = 0.3)
```
